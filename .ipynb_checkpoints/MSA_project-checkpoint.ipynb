```python
modules = [1, 2, 3, 4, 5]
components = ["C1"]
filt = (df["ModuleID"].isin(modules)) & (df["Location Name"].isin(components))
```

```python
df_filtered = df.loc[filt, ["Index", "Location Name", "ModuleID", "OffsetX", "OffsetY", "Comp.Result", 'SizeX', 'SizeY']]
```

```python
pd.set_option('display.max_rows', 72)
df_filtered
```

```python
df_filtered[["OffsetX", "OffsetY"]] = df_filtered[["OffsetX", "OffsetY"]].apply(lambda x: x/1000)
#df_filtered["OffsetY"] = df_filtered["OffsetY"].apply(lambda x: x/1000)
```

```python
df_filtered
```

```python
l = len(df_filtered["OffsetX"])
index = pd.Index(list(range(l)))
df_filtered.index = index
```

```python
df_filtered
```

```python
filt_module = (df_filtered["ModuleID"] == modules[0]) & (df_filtered["Location Name"] == components[0])
insp_counter = len(df_filtered.loc[filt_module, "Index"])
```

```python
a, b = 0, 4
for i in range(1, insp_counter + 1):
    df_filtered.loc[a:b, "Index"] = i
    a += 5
    b += 5
```
